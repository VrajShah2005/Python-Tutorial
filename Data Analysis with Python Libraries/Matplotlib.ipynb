# Matplotlib

```python
import matplotlib.pyplot as plt
```

### Linear Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.plot(x,y)
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = 'r'
plt.plot(x,y,c)
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.plot(x,y)
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.plot(x,y, marker = 'o')
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.plot(x,y)
plt.plot(y, marker = ">")
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.plot(x,y)
plt.plot(y, marker = ">")
plt.plot(x, marker = ">")
```

### Bar Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.bar(x,y)
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']
plt.bar(x,y, color = c)
```

```python
import numpy as np
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3,8,1,10])

plt.bar(x,y, color = "hotpink")
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3,8,1,10])

plt.bar(x,y, color = "#4CAF50")
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3,8,1,10])

plt.bar(x,y, color = "#4CAF50", width = 0.3)
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3,8,1,10])

plt.barh(x,y, color = "#4CAF50", height = 0.1)
```

### Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter plot")
plt.xlabel("month")
plt.ylabel("numbers")

c = ['red', 'magenta', 'green', 'orange', 'blue', 'brown']
plt.scatter(x,y, color = c, s = 150)
plt.show()
```

### Pie Chart

```python
x = [10,20,30,40]
y = ['english', 'hindi', 'maths', 'science']

plt.pie(x, labels = y)
plt.legend()
```
